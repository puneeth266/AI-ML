```python
from sklearn.model_selection import train_test_split

# Split the data into features (X) and response (y)
X = data['area']  
y = data['price']

# Split the data into a training set (80%) and a testing set (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.8, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, explained_variance_score
mse = mean_squared_error(y_test, y_pred)
rmse=np.sqrt(mean_squared_error(y_test, y_pred))
Variance_score=explained_variance_score(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("root-Mean Squared Error:", rmse)
print("Variance_score:", Variance_score)
print("R-square:", r2)
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Metric labels
metrics = ['MSE', 'RMSE', 'Variance Score', 'R-squared']

# Values for both models
values1 = [mse, rmse, Variance_score, r2]
values2 = [mse_M, rmse_M, Variance_score_M, r2_M]

# Set the positions for the bars
x = np.arange(len(metrics))

# Create separate bar charts for each metric
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 6))

# Mean Squared Error (MSE)
axes[0, 0].bar(['single linear regression Model ', 'Multiple linear regression Model'], [values1[0], values2[0]], color=['b', 'r'], alpha=0.7)
axes[0, 0].set_ylabel('Value')
axes[0, 0].set_title('Mean Squared Error')

# Root-Mean Squared Error (RMSE)
axes[0, 1].bar(['single linear regression Model', 'Multiple linear regression Model'], [values1[1], values2[1]], color=['b', 'r'], alpha=0.7)
axes[0, 1].set_ylabel('Value')
axes[0, 1].set_title('Root-Mean Squared Error')

# Variance Score
axes[1, 0].bar(['single linear regression Model', 'Multiple linear regression Model'], [values1[2], values2[2]], color=['b', 'r'], alpha=0.7)
axes[1, 0].set_ylabel('Value')
axes[1, 0].set_title('Variance Score')

# R-squared (R^2)
axes[1, 1].bar(['single linear regression Model', 'Multiple linear regression Model'], [values1[3], values2[3]], color=['b', 'r'], alpha=0.7)
axes[1, 1].set_ylabel('Value')
axes[1, 1].set_title('R-squared')

# Adjust spacing
plt.tight_layout()

# Display the plots
plt.show()
```
