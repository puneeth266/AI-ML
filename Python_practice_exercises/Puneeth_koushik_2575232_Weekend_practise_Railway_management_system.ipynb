```python
import mysql.connector

# Connect to the MySQL database
db = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="Siri@123",
    database="railway_management"
)

# Create a cursor to interact with the database
cursor = db.cursor()

max_seats = 8
def book_ticket(train_id, passenger_name, travel_date, ticket_class):
    try:
        # Check seat availability
        cursor.execute("SELECT COUNT(*) FROM Bookings WHERE TrainID = %s AND TravelDate = %s", (train_id, travel_date))
        count = cursor.fetchone()[0]
        
        if count >= max_seats:
            print("No available seats for the selected train and date.")
            return

        # Insert booking record
        cursor.execute("INSERT INTO Bookings (TrainID, PassengerName, TravelDate, Class, Status) VALUES (%s, %s, %s, %s, %s)",
                       (train_id, passenger_name, travel_date, ticket_class, 'Confirmed'))
        db.commit()
        print("Ticket booked successfully!")

    except Exception as e:
        print("Error:", e)

# Task 2: Reservation Cancellation
def cancel_booking(booking_id):
    try:
        cursor.execute("UPDATE Bookings SET Status = 'Cancelled' WHERE BookingID = %s", (booking_id,))
        db.commit()
        print("Booking cancelled successfully!")

    except Exception as e:
        print("Error:", e)

# Task 3: Fare Checking
def check_fare(train_id, ticket_class):
    try:
        cursor.execute("SELECT FareAmount FROM Fares WHERE TrainID = %s AND Class = %s", (train_id, ticket_class))
        fare = cursor.fetchone()[0]
        print("Fare for the selected class:", fare)

    except Exception as e:
        print("Error:", e)

# Task 4: Booking History
def booking_history(passenger_name):
    try:
        cursor.execute("SELECT * FROM Bookings WHERE PassengerName = %s", (passenger_name,))
        bookings = cursor.fetchall()
        
        if not bookings:
            print("No booking history found for the passenger.")
        else:
            for booking in bookings:
                print("BookingID:", booking[0])
                print("TrainID:", booking[1])
                print("PassengerName:", booking[2])
                print("TravelDate:", booking[3])
                print("Class:", booking[4])
                print("Status:", booking[5])
                print()

    except Exception as e:
        print("Error:", e)

# Task 5: Train Availability
def train_availability(source_station, destination_station):
    try:
        cursor.execute("SELECT * FROM Trains WHERE SourceStation = %s AND DestinationStation = %s",
                       (source_station, destination_station))
        available_trains = cursor.fetchall()
        
        if not available_trains:
            print("No available trains for the selected route.")
        else:
            for train in available_trains:
                print("TrainID:", train[0])
                print("TrainName:", train[1])
                print("DepartureTime:", train[2])
                print("ArrivalTime:", train[3])
                print()

    except Exception as e:
        print("Error:", e)

# Task 6: Menu and User Interaction (Sample Menu)
while True:
    print("Railway Management System")
    print("1. Book Ticket")
    print("2. Cancel Booking")
    print("3. Check Fare")
    print("4. Booking History")
    print("5. Train Availability")
    print("6. Exit")
    
    choice = input("Enter your choice: ")

    if choice == '1':
        train_id = input("Enter TrainID: ")
        passenger_name = input("Enter Passenger Name: ")
        travel_date = input("Enter Travel Date (YYYY-MM-DD): ")
        ticket_class = input("Enter Class: ")
        book_ticket(train_id, passenger_name, travel_date, ticket_class)
    
    elif choice == '2':
        booking_id = input("Enter BookingID to cancel: ")
        cancel_booking(booking_id)

    elif choice == '3':
        train_id = input("Enter TrainID: ")
        ticket_class = input("Enter Class: ")
        check_fare(train_id, ticket_class)

    elif choice == '4':
        passenger_name = input("Enter Passenger Name: ")
        booking_history(passenger_name)

    elif choice == '5':
        source_station = input("Enter Source Station: ")
        destination_station = input("Enter Destination Station: ")
        train_availability(source_station, destination_station)

    elif choice == '6':
        break

# Close the database connection
db.close()
```
